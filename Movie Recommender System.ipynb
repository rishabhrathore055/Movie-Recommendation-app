```python
# genre
# id
# keywords
# title
# overview
# cast 
# crew

movies = movies[['movie_id','title','keywords','genres','overview','cast','crew']]
```

```python
movies.dropna(inplace=True)
```

```python
import ast
```

```python
def arrange(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
arrange('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres'] = movies['genres'].apply(arrange)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(arrange)
movies.head()
```

```python
 #Here We Only collect the top 3 cast from the given cast 
def convert(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
 #Here We Only Want The Director From The Crew
    
def find_director(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L        
```

```python
movies['crew'] = movies['crew'].apply(find_director)
```

```python
movies.head(1)
```

```python
 # Here we typecast the String into List
    
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
# Now We Transform The String Into a Single antity so that the recommender model won't we confused
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
# Now we Add  new Column 'tags' in the movies dataset by concatenate overview,genres,keywords,cast,crew.
movies['tags'] = movies['overview'] + movies['keywords'] + movies['genres'] + movies['cast'] + movies['crew']
```

```python
movies.head(2)
```

```python
# Now create new Dataframe with movie_id,title and tags
new_df = movies[['movie_id','title','tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df['tags'][0]
```

```python
new_df['tags'][1]
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

### Text Vectorization using Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words ='english')
```

```python
cv.fit_transform(new_df['tags']).toarray()
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
#len(cv.get_feature_names())
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse = True,key =lambda x:x[1])[1:7]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse = True,key =lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('avtar')
```
